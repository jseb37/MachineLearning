```python
import array
import os
import scipy.misc
import codecs
import numpy as np
from tqdm import tqdm
import pandas as pd
from sklearn.preprocessing import normalize
import joblib
```

```python
def image_extract_asm():
    for asmfile in os.listdir('F:/asmFiles/'):
        filename = asmfile.split('.')[0]
        file = codecs.open('F:/asmFiles/' + asmfile, 'rb')
        filelength = os.path.getsize('F:/asmFiles/' + asmfile)
        width = int(filelength ** 0.5)
        print(filelength)
        rem = int(filelength / width)
    
        arr = array.array('B')
 
        arr.frombytes(file.read())
        file.close()
        reshaped_array = np.reshape(arr[:width * width], (width, width))

        reshaped_array = np.uint8(reshaped_array)
        
    
        scipy.misc.imsave('F:/asmFiles_image/' + filename + '.png',reshaped_array)
```

```python
image_extract_asm()
```

# First 200 Image Pixels

```python
import cv2
imagefeatures = np.zeros((10868, 200))
```

```python
fnames =[]
for i, asmfile in tqdm(enumerate(os.listdir('F:/asmFiles_image/'))):
    img = cv2.imread('F:/asmFiles_image/' + asmfile.split('.')[0] + '.png')
    img_arr = img.flatten()[:200]
    imagefeatures[i, :] += img_arr
```

```python
np.save('imagefeatures',imagefeatures)
```

```python
with open('imagefeatures.npy', 'rb') as f:
   imagefeatures = np.load(f)
print(imagefeatures)
```

```python
image_features_name = []
for i in tqdm(range(200)):
    image_features_name.append('pixel' + str(i))
imgdf = pd.DataFrame(normalize(imagefeatures), columns = image_features_name)
```
